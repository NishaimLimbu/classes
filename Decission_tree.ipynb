```python
data=pd.read_csv("cars.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
data.columns
```

```python
data1=data.drop(columns=['Brand','Car_ID','Model'])
```

```python
data1["Fuel_Type"]=data1['Fuel_Type'].map({'Petrol':1,'Diesel':0})
data1["Transmission"]=data1['Transmission'].map({'Manual':1,'Automatic':0})
data1['Owner_Type']=data1['Owner_Type'].map({'First':1,'Second':2,'Third':3})
```

```python
data1.head()
```

```python
data.shape
```

```python
x=data1.drop('Price', axis=1)
y=data1["Price"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=23545)
```

```python
model=RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
accuracy=accuracy_score(y_test,y_pred)
print(f" accuracy in percentage is {accuracy * 100:.2f}%")
cm=confusion_matrix(y_test,y_pred)
print (cm)

```
