```python
data = data.drop(columns=['PassengerId','Name','Cabin','Ticket'])
```

```python
data['Embarked']=data['Embarked'].fillna(data['Embarked'].mode()[0])
data.isnull().sum()
```

```python
data["Age"]=data['Age'].fillna(data["Age"].median())
data.isnull().sum()
```

```python
data["Sex"]=data["Sex"].map({'male':0,'female':1})
data.head()
```

```python
data1=data
data1.head()
```

```python
x=data[['Pclass','Age','Fare','Sex']]
y=data['Survived']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
print (f"accuracy in percentage is {accuracy * 100:.2f}%")
```

```python
# to increase the accuracy of model
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model,x,y,cv=8)
print (f"cress_validation scores:{scores}")
print (f"mean accuracy: {scores.mean() * 100:.2f}%")
```

```python
Pclass = int(input("Enter pClass (1, 2 or 3): "))
age = float(input("Enter Age: "))
fare = float(input("Enter Fare: "))
sex_input = input("Enter sex (male/female): ")

# Convert sex into numeric: 1 for female, 0 for male
sex = 1 if sex_input.lower() == "female" else 0

# Create DataFrame with correct feature names (must match model training)
user_data = pd.DataFrame({
    'Pclass': [Pclass],
    'Age': [age],
    'Fare': [fare],   # Correct capitalization
    'Sex': [sex]      # Correct capitalization
})

prediction = model.predict(user_data)
print("\nPrediction:", "Survived" if prediction[0] == 1 else "Not Survived")
```
