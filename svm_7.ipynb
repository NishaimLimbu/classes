```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, classification_report
import matplotlib.pyplot as plt
from sklearn.svm import SVC
```

```python
x=data.drop('gender', axis=1)
y=data["gender"]
```

```python
data['gender']=data['gender'].map({'male':1,'female':0})
```

```python
data.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_scaler=scaler.fit_transform(x)
x_scaler[:5]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=SVC()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
accuracy=accuracy_score(y_test,y_pred)
print(f"accuracy in precent: {accuracy * 100:.2f}")
precision=precision_score(y_test,y_pred,average='macro')
print (f"precision score in percent: {precision * 100:.2f}")
cr=classification_report(y_test,y_pred)
print (cr)
```

```python
# to increase the accuracy of model
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model,x,y,cv=5)
print (f"cress_validation scores:{scores}")
print (f"mean accuracy: {scores.mean() * 100:.2f}%")
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
```

```python
svm_model=SVC()
param_grid={'C':[0.1,1,10], 'kernel':['linear','rbf']}
grid=GridSearchCV(svm_model, param_grid, cv=5)
grid.fit(x_train,y_train)
print ("Best parameters:",grid.best_params_)
print ("Best Score:",grid.best_score_)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
```

```python
scm_model=SVC()
param_dist={'C':randint(0,100), 'kernel':['linear','rbf']}
random_search=RandomizedSearchCV(scm_model, param_distributions=param_dist, n_iter=20,cv=5)
random_search.fit(x_train,y_train)
print ("Best parameters: ",random_search.best_params_)
print ("Best Score: ", random_search.best_score_)
```
