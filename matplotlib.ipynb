```
import matplotlib.pyplot as plt
```

```
x=[1,6,3,9,5,6]
y=[9,8,3,6,7,4]
plt.plot(x,y)
plt.show()
```

```
month=['jan', "feb", "mar", 'apr','may', 'jun','july', 'aug','sep','oct','nov','dec']
profit=[12000,11000,8000,8900,11000,11540,12000,14000,10000,15000,14500,13500]
plt.plot(month,profit)
plt.show()
```

```
month=['jan', "feb", "mar", 'apr','may', 'jun','july', 'aug','sep','oct','nov','dec']
profit=[12000,11000,8000,8900,11000,11540,12000,14000,10000,15000,14500,13500]
plt.plot(month,profit, label="line Plot")
plt.xlabel("Months")
plt.title("Earning of the year")
plt.ylabel("Profit")
plt.legend()
plt.show()
```

```
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D', 'E']
values = [20, 15, 18, 20, 25]

plt.bar(categories, values, color="red")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Plot Example")
plt.show()
```

```
x=[1,2,3,4,5,6]
y=[4,8,6,9,2,10]
plt.scatter(x,y,color='green')
plt.xlabel("X-axis")
plt.ylabel("y-axis")
plt.title("Scatter plot example")
plt.show()
```

```
data=[1,2,3,3,4,9,7,6,9,8,4,5,3,3,1,3,3,2,1,5,4,3,5,6,4,2,9,7,6,4]
plt.hist(data,bins=6,color='brown')
plt.xlabel("Data variable")
plt.ylabel("Frequency")
plt.title("Histogram example")
plt.show()
```

```
x=[1,2,3,4,5]
y=[6,7,8,6,4]
plt.plot(x,y,color='brown', marker='o', label='line plot', linestyle='--')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.title("Design example")
plt.show()
```
