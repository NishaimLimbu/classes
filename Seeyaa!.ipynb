```python
from sklearn.preprocessing import LabelEncoder

categorical_cols = ['protocol_type', 'service', 'flag'] 

for col in categorical_cols:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col].astype(str))  # convert to string to avoid errors

# Separate encoder for gender
le_class = LabelEncoder()
data['class'] = le_class.fit_transform(data['class'].astype(str))
```

```python
data.head(5)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
from sklearn.ensemble import IsolationForest
from sklearn.svm import OneClassSVM
```

```python
data.head(1)
```

```python
x=data.drop(['class'],axis=1)
y=data['class']
```

```python
scaler=StandardScaler()
x_scaler=scaler.fit_transform(x)
x_scaler
```

```python
data['class'].value_counts()
```

```python
iso_forest = IsolationForest(contamination=0.46614, random_state=434)
```

```python
data['anomaly']=iso_forest.fit_predict(x)
```

```python
data['anomaly'].value_counts()
```

```python
data['anomaly']=data['anomaly'].map({1:int(1),-1:int(0)})
data.head(1)
```
