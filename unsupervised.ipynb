```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
np.random.seed(42)
```

```python
age1=np.random.randint(18,25,50)
income1=np.random.randint(16000,25000,50)

age2=np.random.randint(27,39,50)
income2=np.random.randint(26000,45000,50)

age3=np.random.randint(40,61,50)
income3=np.random.randint(55000,75000,50)
```

```python
ages=np.concatenate([age1,age2,age3])
incomes=np.concatenate([income1,income2,income3])

customer_data=pd.DataFrame({'Age':ages,'Income':incomes})
x=customer_data[['Age','Income']].values
```

```python
from sklearn.cluster import KMeans
wcss=[]
for k in range(1,11):
    kmeans =KMeans(n_clusters=k,random_state=4)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)

# plot the elbow graph
plt.plot(range(1,11),wcss,marker='o')
plt.title("elbow method to find k")
plt.xlabel("no of cluster(k)")
plt.ylabel('wcss')
plt.show()
```

```python
kmeans=KMeans(n_clusters=3,random_state=4)
kmeans.fit(x)

plt.scatter(x[:,0],x[:,1], c=kmeans.labels_, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='red',marker='X')
plt.title("Scatter")
plt.xlabel("Ages")
plt.ylabel("Incomes")
plt.show()
```
